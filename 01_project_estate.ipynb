# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

## Изучение данных из файла

```python
import pandas as pd
import matplotlib.pyplot as plt

real_estate_data = pd.read_csv('/datasets/real_estate_data.csv', sep='\t')
real_estate_data.head()
```

```python
real_estate_data.info()
```

Приведем названия столбцов к нижнему регистру

```python
real_estate_data = real_estate_data.rename(str.lower, axis='columns')
real_estate_data.columns
```

Для удобства просмотра таблицы создадим набор колонок в нужном нам порядке и датафрейм, которым можем пользоваться для визуального анализа

```python
useful_cols = ['locality_name', 'total_area', 'living_area', 'kitchen_area', 
              'studio', 'open_plan', 'rooms', 'floor', 'floors_total', 'ceiling_height',
               'last_price', 'first_day_exposition', 'days_exposition',
               'citycenters_nearest'
              ]
real_estate_data_short = real_estate_data[useful_cols] 
real_estate_data_short.head(5)
```

### Вывод 
В выгрузке присутствуют квартиры разных типов - свободной планировки (не имеющие выделенной площади кухни), студии и апартаменты, что необходимо учитывать при оценке рыночной стоимости. 

На момент выгрузки данных многие квартиры не проданы, соответственно, мы не можем оценивать адекватность выставленной цены.

В таблице также имеются многочисленные артефакты, пропуски и несоответствия, например потолки высотой свыше пятидесяти метров, этаж квартиры выше максимального этажа в доме, сумма жилой площади и кухни, превышающая общую площадь и т.д.

## Предобработка данных

Рассмотрим пропуски в данных

```python
real_estate_data.isnull().sum()
```

Для нашего исследования необходимо заполнить неопределенные данные в столбцах:

-жилая площадь и площадь кухни     
-высота потолков    
-количество комнат    
-населенный пункт   
-расстояние от центра    


1. Столбец `days_exposition` - сколько дней было размещено объявление (от публикации до снятия)

13% квартир не имеют значений в данном столбце, что позволяет предположить, что квартира на момент выгрузки не продана.
Удалим эти строки, так как для расчета рыночной стоимости они не показательны. 

```python
real_estate_data = real_estate_data.dropna(subset=['days_exposition'])
```

2. `balcony` - число балконов

Отсутствующие балконы заполним нулями. 

```python
real_estate_data['balcony'] = real_estate_data['balcony'].fillna(value=0)
```

`citycenters_nearest`    
`airports_nearest`    
`parks_around3000`    
`parks_nearest`    
`ponds_around3000`    
`ponds_nearest`    

Данные по удаленности от центра, паркам и аэропортам присутствуют только для Санкт-Петербурга и ближайших пригородов, так что просто заполним их значением -1, чтобы отличить от значения 0 (парки отсутствуют)

```python
real_estate_data['airports_nearest'] = real_estate_data['airports_nearest'].fillna(value=-1)
real_estate_data['citycenters_nearest'] = real_estate_data['citycenters_nearest'].fillna(value=-1)
real_estate_data['parks_around3000'] = real_estate_data['parks_around3000'].fillna(value=-1)
real_estate_data['parks_nearest'] = real_estate_data['parks_nearest'].fillna(value=-1)
real_estate_data['ponds_around3000'] = real_estate_data['ponds_around3000'].fillna(value=-1)
real_estate_data['ponds_nearest'] = real_estate_data['ponds_nearest'].fillna(value=-1)
```

`is_apartment`

Все, что кроме True, заполним False

```python
real_estate_data['is_apartment'] = real_estate_data['is_apartment'].fillna(value=False)
```

`ceiling_height`

Высота потолков зависит от типа квартиры и обычно составляет от 2.4 до 4 метров. 

Если бы у нас стояла задача исследовать зависимость цены от высоты потолков, стоило бы использовать срез данных только с заполненной информацией о потолках. 
Но так как мы просто рассматриваем этот параметр, заполним его медианой. 


```python
ceil_median = real_estate_data['ceiling_height'].median()
real_estate_data['ceiling_height'] = real_estate_data['ceiling_height'].fillna(value=ceil_median)
```

`floors_total` - всего этажей в доме

Просто заполним текущим этажом+1, чтобы квартира не осталась на последнем этаже, так это может исказить исследование

```python
real_estate_data['floors_total'] = real_estate_data['floors_total'].fillna(value=real_estate_data['floor']+1)
```

`locality_name` - название населённого пункта

Используем данные об удаленности от центра
Населенный пункт можно вычислить, исходя из расстояния от центра города. Из сводных данных видно, что сам ближайший пригород находится на расстоянии 22 км, соответственно, квартиры внутри этого радиуса можно считать находящимися в Санкт-Петербурге.
Заполним столбцец значением "нет данных", а радиус 22 км значением "Санкт-Петербург"

```python
(real_estate_data
.query('citycenters_nearest > 0')
.pivot_table(index='locality_name', values= 'citycenters_nearest', aggfunc = 'max')
.reset_index()
.sort_values(by = 'citycenters_nearest', ascending = True).head(1)
)
```

```python
#real_estate_data.isnull(real_estate_data['locality_name'])] #

real_estate_data['locality_name'] = real_estate_data['locality_name'].fillna(value='нет данных')
real_estate_data.loc[(real_estate_data['citycenters_nearest'] <= 22000) & (real_estate_data['locality_name'] == 'нет данных'), 'locality_name'] = 'Санкт-Петербург' 
```

`kitchen_area`- площадь кухни

Площадь кухни отсутствует для квартир-студий и квартир открытой планировки. Заменим пропуски нулями. 

Для квартир обычного типа площадь кухни заполним, рассчитав ее пропорционально общей площади жилья. Для чего найдем по таблице отношение кухни к общей площади и возьмем медиану от этих значений. 

```python
real_estate_data[pd.isnull(real_estate_data['kitchen_area'])]

real_estate_data.loc[(real_estate_data['open_plan'] == True) | (real_estate_data['studio'] == True), 'kitchen_area'] = 0 


real_estate_data['ktch_to_total_area'] = real_estate_data['kitchen_area']/real_estate_data['total_area']
kta_median = real_estate_data['ktch_to_total_area'].median()
kta_median
real_estate_data['kitchen_area'] = real_estate_data['kitchen_area'].fillna(value=real_estate_data['total_area']*kta_median)

real_estate_data['ktch_to_total_area'] = real_estate_data['kitchen_area']/real_estate_data['total_area']
```

`living_area` - жилая площадь. 

Пропуски в жилой площади также найдем как отношение известной жилой площади к общей и возьмем медиану. 

```python
real_estate_data['lvg_to_total_area'] = real_estate_data['living_area']/real_estate_data['total_area']
lta_median = real_estate_data['lvg_to_total_area'].median()
lta_median

real_estate_data['living_area'] = real_estate_data['living_area'].fillna(value=real_estate_data['total_area']*lta_median)

real_estate_data['lvg_to_total_area'] = real_estate_data['living_area']/real_estate_data['total_area']
```

`rooms` - количество комнат

Для квартир-студий количество комнат 0 заменим на 1, так как по факту одна комната у них есть. 
Аналогично для квартир без планировки.

```python
real_estate_data.loc[(real_estate_data['studio'] == True) & (real_estate_data['rooms'] == 0), 'rooms'] = 1
real_estate_data.loc[(real_estate_data['open_plan'] == True) & (real_estate_data['rooms'] == 0), 'rooms'] = 1
```

`ceiling_height` - высота потолков

Нормальной высотой потолков считается высота от двух до четырех метров.
Можно заметить, что в таблице несколько строк имеют значение "27". Будем считать это ошибкой оператора, и что значению 27 метров на деле соответствует 2,7 метра.

Остальные аномальные значения оставим как есть. 


```python
(real_estate_data
.query('ceiling_height > 6')
.pivot_table(index='ceiling_height', values= 'total_area', aggfunc='count')
)
```

```python
real_estate_data.loc[(real_estate_data['ceiling_height'] > 22), 'ceiling_height'] = real_estate_data['ceiling_height']/10
```

`locality_name`

Удаление дубликатов в населенных пунктах. 

В таблице присутствуют дубликаты названий населенных пунктов, например "посёлок Мурино" и "поселок Мурино", "поселок городского типа Рахья" и "садовое товарищество Рахья". 

Как минимум стоит заменить ё на е в слове "поселок".

```python
real_estate_data['locality_name'] = real_estate_data['locality_name'].str.replace('ё','е')
```

Замена типа данных. 

В связи с избыточностью вещественный тип данных заменен на целочисленный.

```python
real_estate_data['last_price'] = real_estate_data['last_price'].astype('int')
real_estate_data['total_area'] = real_estate_data['total_area'].astype('int')
real_estate_data['balcony'] = real_estate_data['balcony'].astype('int')
real_estate_data['airports_nearest'] = real_estate_data['airports_nearest'].astype('int')
real_estate_data['floors_total'] = real_estate_data['floors_total'].astype('int')
real_estate_data['citycenters_nearest'] = real_estate_data['citycenters_nearest'].astype('int')
real_estate_data['parks_around3000'] = real_estate_data['parks_around3000'].astype('int')
real_estate_data['parks_nearest'] = real_estate_data['parks_nearest'].astype('int')
real_estate_data['ponds_around3000'] = real_estate_data['ponds_around3000'].astype('int')
real_estate_data['ponds_nearest'] = real_estate_data['ponds_nearest'].astype('int')
real_estate_data['days_exposition'] = real_estate_data['days_exposition'].astype('int')
real_estate_data['living_area'] = real_estate_data['living_area'].astype('int')
real_estate_data['kitchen_area'] = real_estate_data['kitchen_area'].astype('int')
real_estate_data['lvg_to_total_area'] = real_estate_data['lvg_to_total_area'].astype('int')
real_estate_data['ktch_to_total_area'] = real_estate_data['ktch_to_total_area'].astype('int')
```

### Вывод
Пропущенные значения обработаны и заполнены, заменен тип данных. 

Основной причиной появления неопределенных и аномальных данных будем считать некорректную выгрузку из разных источников, недочеты в способе сбора информации и просто технические нюансы обработки данных.

## Расчёты и добавление результатов в таблицу

Рассчитаем и добавим в таблицу необходимые дополнительные столбцы. 

- Цена квадратного метра

```python
#цену разделить на площадь квартиры

real_estate_data['sq_meter_price'] = real_estate_data['last_price']/real_estate_data['total_area']
real_estate_data['sq_meter_price'] = real_estate_data['sq_meter_price'].astype('int')
```

- День недели, месяц и год публикации объявления

```python
real_estate_data['weekday'] = real_estate_data['first_day_exposition'].dt.weekday
real_estate_data['month'] = real_estate_data['first_day_exposition'].dt.month
real_estate_data['year'] = real_estate_data['first_day_exposition'].dt.year
```

- Этаж квартиры; варианты — первый, последний, другой

Для упрощения построения графиков разделим все квартиры на три числовых типа: 1 - первый этаж, 3 - последний этаж, 2 - все остальные. 

Аномальные этажи, которые выше, чем число этажей во всем доме, так же относятся к типу "другой". 


```python
#Cоздадим функцию 
def add_floor_type(row):
 
    floor = row['floor']
    floors_total = row['floors_total']
    
    if (floor == floors_total):
        return 'последний'    
    if (floor == 1):
        return 'первый'        
    return 'другой' 

#создадим новый столбец floor_type
real_estate_data['floor_type'] = real_estate_data.apply(add_floor_type, axis=1)

#real_estate_data['floor_type'].value_counts()
real_estate_data['floor'].value_counts() 
```

```python
def floor_index(row):
 
#    floor = row['floor']
#    floors_total = row['floors_total']
    
    if row['floor'] == row['floors_total']:
        return 'последний'    
    elif row['floor'] == 1:
        return 'первый'        
    return 'другой' 

#создадим новый столбец floor_type
real_estate_data['floor_type'] = real_estate_data.apply(floor_index, axis=1)
```

- Цена в млн руб

```python
real_estate_data['last_price_mln'] = real_estate_data['last_price']/1000000
real_estate_data['last_price_mln'] = real_estate_data['last_price_mln'].astype('int')
```

- Расстояние от центра в километрах и средняя стоимость километра

```python
#столбец с расстоянием до центра в километрах
#если расстояния нет, заполним значением -1 
real_estate_data['citycenters_km'] = real_estate_data['citycenters_nearest']/1000
real_estate_data['citycenters_km'] = real_estate_data['citycenters_km'].astype(int)
real_estate_data.loc[(real_estate_data['citycenters_km'] == 0 ),'citycenters_km'] = -1
```

- Cредняя цена для каждого километра

```python
real_estate_data['avg_km_price'] = real_estate_data['last_price']/real_estate_data['citycenters_km'] 
real_estate_data['avg_km_price'].value_counts()
real_estate_data['avg_km_price'] = real_estate_data['avg_km_price'].astype('int')
```

```python
#Cоздадим функцию 
def add_loc_type(row):
    
    if row['citycenters_km'] <= 3:
        return 'центр'    
    if row['locality_name'] == 'Санкт-Петербург' and row['citycenters_km'] > 3:
        return 'город'        
    return 'область' 

real_estate_data['loc_type'] = real_estate_data.apply(add_loc_type, axis=1)

real_estate_data['loc_type'].value_counts()

```

### Вывод

Высота потолков также имеет значение. Причем за пределами центральной части города, в районах, где больше типового жилья, высокие потолки ценятся заметно выше. 

Расстояние от центра так же имеет обратную зависимость, но не столь сильно выраженную. По мере удаления от центра цена квартиры все больше падает. 

Этаж и месяц размещения объявления практически не связаны с ценой, коэффициент приближается к нулю.
